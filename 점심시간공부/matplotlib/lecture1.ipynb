```python
import matplotlib.pyplot as plt
import numpy as np
```

## Subplot으로 추가하기

```python
fig = plt.figure(figsize=(7,7))
ax1 = plt.subplot(2,2,1, facecolor = 'r')
ax2 = plt.subplot(2,2,2, facecolor = 'g')
ax3 = plt.subplot(2,2,3, facecolor = 'b')
ax4 = plt.subplot(2,2,4, facecolor = 'orange')
```

## add_subplot으로 추가하기

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
```

## subplots로 그리기

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot([1,2,3])
axes[1].plot([2,2,2])
```

```python
fig, (ax1,ax2) = plt.subplots(2,1,figsize=(5,3))
ax1.plot([1,2,3])
ax2.plot([2,2,2])
```

다수의 plot을 그릴때 유리

## plt.subplot2grid활용하기

```python
fig = plt.figure()
ax1 = plt.subplot2grid((2,1),(0,0),fig=fig)
ax2 = plt.subplot2grid((2,1),(1,0),fig=fig)
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,2),(0,0),fig=fig)
ax2 = plt.subplot2grid((1,2),(0,1),fig=fig)
```

```python
fig = plt.figure(figsize=(7,7))
ax1 = plt.subplot2grid((3,3),(0,0),
                       colspan=2, fig=fig)

ax2 = plt.subplot2grid((3,3),(1,0),colspan=3,fig=fig)

ax3 = plt.subplot2grid((3,3),(2,0),fig=fig)
ax4 = plt.subplot2grid((3,3),(2,1),fig=fig)
ax5 = plt.subplot2grid((3,3),(2,2),fig=fig)
```

```python
fig = plt.figure(figsize=(7,7),facecolor='skyblue')
ax1 = plt.subplot2grid((2,2),(0,0),colspan=2,fig=fig)
ax2 = plt.subplot2grid((2,2),(1,0),fig=fig)
ax2 = plt.subplot2grid((2,2),(1,1),fig=fig)
```

```python
fig = plt.figure(figsize=(7,7),facecolor='skyblue')
ax1 = plt.subplot2grid((3,3),(0,0),colspan=2, fig=fig)
ax2 = plt.subplot2grid((3,3),(1,0),colspan=2, fig=fig)
ax3 = plt.subplot2grid((3,3),(2,0),colspan=2, fig=fig)
ax4 = plt.subplot2grid((3,3),(0,2),rowspan=3, fig=fig)
```

```python
fig = plt.figure(figsize=(7,7),facecolor='skyblue')
ax1 = plt.subplot2grid((3,3),(0,0),colspan=3,fig=fig)
ax2 = plt.subplot2grid((3,3),(1,0),colspan=2,fig=fig)
ax3 = plt.subplot2grid((3,3),(1,2),rowspan=2,fig=fig)
ax4 = plt.subplot2grid((3,3),(2,0),fig=fig)
ax5 = plt.subplot2grid((3,3),(2,1),fig=fig)
```

## add_axes활용하기

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure(figsize=(7,7),facecolor='skyblue')
ax = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
fig = plt.figure(figsize=(7,7),facecolor='skyblue')
ax1 = fig.add_axes([0.1,0.1,0.5,0.5])
ax1 = fig.add_axes([0.7,0.1,0.2,0.5])
ax1 = fig.add_axes([0.1,0.7,0.8,0.2])
```

```python
fig = plt.figure(figsize=(7,7),facecolor='skyblue')
rect1 = [0.1,0.1,0.5,0.5]
rect2 = [0.7,0.1,0.2,0.5]
rect3 = [0.1,0.7,0.8,0.2]

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```
