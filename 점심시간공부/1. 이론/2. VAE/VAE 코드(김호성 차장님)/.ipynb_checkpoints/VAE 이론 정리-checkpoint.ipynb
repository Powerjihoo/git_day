# VAE(Variational Autoencoder) 개요

VAE는 생성 모델 중 하나로, 데이터의 잠재 표현을 학습하고, 이를 사용하여 새로운 데이터를 생성하는 모델. 

VAE는 오토인코더(Autoencoder)의 확장된 형태로, 잠재 공간의 구조를 가우시안 분포에 기반하여 모델링함.

VAE는 **인코더(encoder)와 디코더(decoder)**라는 두 개의 신경망으로 구성.


## VAE의 핵심 아이디어

잠재 공간(latent space)에서의 분포를 학습하여 데이터를 생성하는 확률적인 모델.  
  
인코더는 입력 데이터를 저차원의 잠재 변수로 인코딩하고, 디코더는 잠재 변수를 원래의 데이터 공간으로 디코딩.  

잠재 변수는 일반적으로 가우시안 분포(정규 분포)를 따르는 것으로 가정되며, 인코더는 입력 데이터를 이 분포의 파라미터로 매핑.  

디코더는 잠재 변수로부터 원래의 데이터 분포를 복원하는 역할을 수행.  


## VAE의 학습 방법

학습 과정에서 변분 추론(variational inference)을 사용하여 잠재 변수를 학습합니다. 

이는 입력 데이터와 잠재 변수 간의 관계를 추론하는 것으로, 주어진 입력 데이터에 대한 잠재 변수의 사후 분포를 근사화하는 과정입니다. 


인코더를 통해 입력 데이터를 잠재 공간으로 매핑하고, 잠재 변수의 **평균과 분산**을 추정합니다.  


추정된 평균과 분산을 사용하여 잠재 변수를 샘플링합니다.  


디코더를 사용하여 잠재 변수를 원본 입력으로 복원합니다.  


복원된 입력과 원본 입력 간의 차이를 최소화하기 위해, 재구성 손실(reconstruction loss)을 계산. 일반적으로 원본 입력과 복원된 입력의 **평균 제곱 오차(Mean Squared Error, MSE)**로 정의.  


잠재 변수의 사후 분포와 사전 분포 사이의 차이를 최소화하기 위해, KL divergence을 계산하여 정규화 손실(regularization loss)을 추가.

재구성 손실과 정규화 손실을 합하여 총 손실(total loss)를 계산.  


총 손실을 최소화하는 방향으로 모델의 가중치를 업데이트.  



## VAE의 구조

VAE는 인코더와 디코더로 구성.  

 - 인코더(Encoder): 입력 데이터를 저차원의 잠재 공간(latent space)으로 매핑합니다. 인코더는 주어진 입력 데이터를 잠재 변수의 평균과 분산으로 매핑하는 인코딩 함수로 구성됩니다.  

 - 디코더(Decoder): 잠재 공간의 저차원 벡터를 원본 입력 데이터로 복원합니다. 디코더는 잠재 변수에서 원본 입력을 복원하는 디코딩 함수로 구성됩니다.  


VAE에서는 일반적으로 인코더와 디코더 모두 다층 퍼셉트론(MLP) 구조를 사용.  


인코더와 디코더 사이에는 평균과 분산을 처리하는 MLP가 하나씩 더 존재.  




  




## VAE의 활용

- VAE는 데이터의 잠재적인 특성을 학습하고, 새로운 데이터를 생성하는 능력을 가지고 있습니다.   

- 이러한 특성으로 인해 VAE는 다양한 응용 분야에서 사용될 수 있습니다.   

 - 예를 들어, 이상 탐지(anomaly detection)에서는 VAE를 사용하여 정상 데이터의 분포를 학습하고, 이상치로 판단되는 데이터를 식별합니다. 또한, 이미지 생성, 음악 생성, 자연어 처리 등 다양한 창작적인 작업에도 활용될 수 있습니다.  

# 코드 예시  - VAE, 베이스

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
from keras.layers import Layer, Dense, Flatten, Reshape, Input
from keras.models import Model
from keras.optimizers import Adam
import pandas as pd
import matplotlib.pyplot as plt

from keras import backend as K
mse_loss_fn = tf.keras.losses.MeanSquaredError()
optimizer = Adam()
```

```python
def calc_kl_loss(z_mean, z_log_var, weight=0.0001):
    loss_kl = 1 + z_log_var - K.square(z_mean) - K.exp(z_log_var)
    loss_kl = K.mean(loss_kl)
    loss_kl *= -0.5
    loss_kl = weight * loss_kl
    return loss_kl
```

```python
class Sampling(Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = keras.backend.random_normal(shape=(batch,dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon
```

```python
class VAE(Model):
    def __init__(self, window_size:int, feature_dim:int):
        super(VAE, self).__init__()
        self.inputs = Flatten(name="Input")
        self.encoder_dense1 = Dense(units=32, name="Encoder_Dense_1", activation="relu")
        self.encoder_dense2 = Dense(units=16, name="Encoder_Dense_2", activation="relu")
        self.z_mean = Dense(units=8, name="Encoder_Dense_mean")
        self.z_log_var = Dense(units=8, name="Encoder_Dense_log_var") 
        self.sampling = Sampling()
        self.decoder_dense1 = Dense(units=16, name="Decoder_Dense_1", activation="relu")
        self.decoder_dense2 = Dense(units=feature_dim*window_size, name="Decoder_Dense_Final", activation="relu")
        self.outputs = Reshape(target_shape=(window_size, feature_dim), name="Output")
    
    def call(self, x):
        x = self.inputs(x)
        x = self.encoder_dense1(x)
        x = self.encoder_dense2(x)
        
        z_mean = self.z_mean(x)
        z_log_var = self.z_log_var(x)
        z = self.sampling((z_mean, z_log_var))
        
        x = self.decoder_dense1(z)
        x = self.decoder_dense2(x)
        x = self.outputs(x)
        
        return x, z_mean, z_log_var
```

```python
@tf.function
def train_step():
    for data_batch in dataset_train:
        with tf.GradientTape() as tape:
            # Forward propagation
            preds, z_mean, z_log_var = model(data_batch)
            
            # Loss function
            recons_loss = K.mean(mse_loss_fn(data_batch, preds))
            kl_loss = 1 + z_log_var - K.square(z_mean) - K.exp(z_log_var)
            kl_loss = K.mean(kl_loss)
            kl_loss *= -0.5
            kl_loss = 0.0001 * kl_loss
            total_loss = recons_loss + kl_loss

            # Backpropagation
            grads = tape.gradient(total_loss, model.trainable_variables)
            optimizer.apply_gradients(zip(grads, model.trainable_variables))
        # print(total_loss.numpy(), recons_loss.numpy(), kl_loss.numpy())
```

```python
def train_model(epochs=5):
    for epoch in range(epochs):
        print(f"Epoch: {epoch}")
        train_step()
```

## 베르누이 분포(Bernoulli Distribution)

베르누이 분포(Bernoulli distribution)는 이항 분포(Binomial distribution)의 특수한 경우로, 이항 분포에서 시행 횟수가 1인 경우를 나타냅니다. 즉, 베르누이 분포는 이항 분포에서 단 하나의 시행만을 수행할 때의 확률 분포입니다.

베르누이 분포는 이항 변수(0 또는 1)의 확률 분포를 나타내며, 이항 변수의 값이 1일 확률을 p로 나타냅니다. 

베르누이 분포는 이항 변수의 값이 1일 확률이 p이고, 0일 확률이 1-p인 이항 분포로 간주될 수 있습니다.

베르누이 분포는 이진 분류 문제(binary classification)에서 사용되며, 성공 또는 실패와 같은 두 가지 이벤트를 모델링하는데 유용합니다. 예를 들어, 동전 던지기에서 앞면이 나올 확률을 p로 가정한다면, 동전 던지기 결과는 베르누이 분포를 따를 것입니다.

머신 러닝에서 베르누이 분포는 로지스틱 회귀(Logistic Regression)와 같은 분류 모델에서 활용되며, 이진 분류 문제의 확률 예측에 사용될 수 있습니다. 또한, 베르누이 분포는 생성 모델(Generative Model)에서 확률적인 데이터 생성을 모델링하는데에도 사용됩니다.

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import bernoulli
```

```python
p = 0.3  # 성공 확률

x = [0, 1]  # 가능한 값
y = bernoulli.pmf(x, p)  # 확률 질량 함수 (PMF)

plt.bar(x, y)
plt.xticks(x)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution (p=0.3)')
plt.show()
```

# 가우시안 분포(Gaussian Distribution = 정규 분포)

가우시안 분포(Gaussian distribution)는 통계학에서 가장 잘 알려진 확률 분포 중 하나입니다. 정규 분포(Normal distribution)라고도 불리며, 연속 확률 분포의 한 종류입니다. 가우시안 분포는 평균(Mean)과 표준 편차(Standard deviation)라는 두 개의 매개 변수로 정의됩니다.

가우시안 분포는 평균을 중심으로 대칭적인 종 모양을 가지며, 표준 편차가 작을수록 데이터가 평균 주위에 모여있는 형태를 가집니다.

가우시안 분포는 다양한 자연 현상을 모델링하는 데에 널리 사용됩니다. 예를 들어, 사람의 키, 체중, IQ 점수 등은 보통 가우시안 분포에 근사적으로 따르는 경우가 많습니다. 또한, 중심 극한 정리에 의해 여러 독립적인 확률 변수들의 합은 가우시안 분포에 근사하게 됩니다.

데이터 분석, 통계 모델링, 머신 러닝 등 다양한 분야에서 가우시안 분포는 중요한 개념으로 활용됩니다. 특히, 가우시안 분포의 특성을 이용하여 데이터의 패턴을 분석하고, 이상치 탐지, 분류, 예측 등의 작업을 수행하는데 활용될 수 있습니다.

```python
mu = 0  # 평균
sigma = 1  # 표준편차

x = np.linspace(-5, 5, 100)  # x 범위
y = norm.pdf(x, mu, sigma)  # 확률 밀도 함수 (PDF)

plt.plot(x, y)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Gaussian Distribution (mu=0, sigma=1)')
plt.show()
```

# Reparamenterization trick = 샘플링하는 과정

- 평균 및 분산 활용

- 잠재 변수 분포 정의, 정규 분포 사용
 
- 랜덤 노이즈 생성

- 잠재 변수 역전파 가능하게 함

일반적으로, 잠재 변수 z의 분포를 평균 μ와 표준 편차 σ로 표현할 수 있습니다. 그런 다음, 정규 분포 N(0, 1)에서 샘플 ε을 뽑아서 잠재 변수 z를 다음과 같이 계산:

z = μ + ε * σ

여기서 ε은 평균이 0이고 표준 편차가 1인 정규 분포에서 샘플링한 값입니다. 이렇게 잠재 변수 z를 샘플링하는 과정을 통해 역전파를 가능하게 하고, 잠재 변수의 확률적인 특성을 보존하면서 학습이 가능

아래는 예시 코드

```python
def reparameterize(self, mu, logvar):
    epsilon = torch.randn_like(mu)
    std = torch.exp(0.5 * logvar)
    z = mu + epsilon * std
    return z
```

- VAE의 인코더를 통해 얻은 평균(mu)과 로그 분산(logvar)을 사용하여 잠재 변수(z)를 재매개화하는 역할.

- epsilon을 무작위로 생성, 표준편차(std)는 로그 분산을 통해 계산

- 잠재 변수(z)는 평균에 표준편차를 곱한 후 epsilon을 곱하여 얻음

# regularization loss(정규화 손실) & reconstruction loss(재구성 손실)

 정규화 손실은 VAE의 잠재 공간(z)을 정규화하고 잠재 변수의 분포를 조절하기 위해 사용. 일반적으로 VAE는 잠재 공간을 가우시안 정규분포에 가깝게 만들기 위해 정규화 손실을 최소화하는 방향으로 학습. 이를 위해 VAE 모델은 잠재 변수(z)의 평균(mu)과 로그 분산(logvar)을 출력하는 인코더를 사용하고, 정규화 손실은 평균과 분산의 KL divergence(Kullback-Leibler divergence)로 계산. KL divergence는 실제 분포와 모델이 가정한 분포 사이의 차이를 측정하는 지표로 사용. 정규화 손실은 잠재 변수의 분포를 가우시안에 가깝게 유지하도록 모델을 규제하고, 잠재 공간의 구조를 학습하는 데 도움을 줌.
 
 재구성 손실은 VAE의 디코더를 통해 입력 데이터를 재구성하는 과정에서 발생하는 손실을 의미. VAE는 입력 데이터를 잠재 변수(z)로 인코딩한 후, 디코더를 사용하여 잠재 변수를 다시 원래의 입력 데이터로 디코딩. 재구성 손실은 입력 데이터와 재구성된 데이터 간의 차이를 측정하여 계산됩니다. 일반적으로 VAE는 재구성 손실을 최소화하는 방향으로 학습하여 입력 데이터를 정확하게 재구성할 수 있도록 함. 재구성 손실은 평균 제곱 오차(Mean Squared Error, MSE)나 이진 교차 엔트로피(Binary Cross Entropy, BCE) 등을 사용하여 계산 가능.

# 의문점 정리

### 인코더에서 나온 출력 값에서 평균과 분산을 받는 MLP는 어떻게 값이 달라지나?

- 순전파, 역전파에 의해서 가중치 및 편향이 조정되며 평균과 분산에 가까운 값으로 바뀜

### 코드에 self가 계속 나오는 이유

- 외부 변수를 참조하지 않고, 클래스 내부 인스턴스 및 메서드에 접근하기 위함

### 정규화 손실

- 모델 가중치 제한, 과적합 방지

### 재구성 손실

- 입력 데이터와 재구성 데이터의 차이. MSE, Binary Cross Entropy 사용

### total loss

- 위 두 개의 loss를 활용하여 나온 지표. 단순 덧셈 또는 가중치를 부여하여 여러 방법으로 구할 수 있음
