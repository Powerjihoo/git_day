```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
import warnings
warnings.filterwarnings('ignore')
```

## 데이터 불러오기 및 확인

```python
use_data = pd.read_csv('simulation/total_data.csv', parse_dates=['_time'], index_col='_time')[:100]
model = ARIMA(use_data[:60], order=(12, 2, 6))
model_fit = model.fit()

forecast = model_fit.forecast(steps=30)
forecast_df = pd.DataFrame(forecast)

# 그래프 그리기
plt.figure(figsize=(23, 8))
plt.plot(use_data[:90]['_value'].values, label='original data', color='orange')
plt.axvline(60,color='red')

start_index = 60
end_index = start_index + len(forecast_df['predicted_mean'])
plt.title("ARIMA Prediction vs Original Data")
plt.plot(range(start_index, end_index), forecast_df['predicted_mean'], label='arima forecast', color='blue')
plt.grid()
plt.legend()
plt.show()
```

## 예측하기(Prophet)

```python
use_data = pd.read_csv('simulation/total_data.csv')[:100]
use_data = use_data.rename(columns={'_time': 'ds', '_value': 'y'})
use_data['ds'] = pd.to_datetime(use_data['ds']).dt.tz_localize(None)


model = Prophet()
model.add_seasonality(name='12min_cycle', period=12, fourier_order=6)
model.fit(use_data[:60])

future = model.make_future_dataframe(periods=30, freq='T')  # 60분 예측
forecast = model.predict(future)

# 그래프 그리기
plt.figure(figsize=(23, 8))
plt.plot(use_data[:90]['y'].values, label='original data', color='orange')
plt.axvline(60,color='red')
start_index = 60
end_index = start_index + len(forecast[60:]['yhat'])
plt.plot(range(start_index, end_index), forecast[60:]['yhat'], label='prophet forecast', color='blue')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Dropout

# 데이터 로드 및 준비
# _value 열만 사용, 데이터는 CSV 파일로 가정
data = pd.read_csv("simulation/total_data.csv", parse_dates=["_time"])
values = data["_value"].values  # numpy array로 변환

# 하이퍼파라미터
sequence_length = 60  # 입력 길이 (60분)
predict_length = 30   # 예측 길이 (30분)
batch_size = 16
hidden_units = 16
num_epochs = 120

# 데이터 윈도우 생성 함수
def create_sequences(data, input_length, output_length):
    x, y = [], []
    for i in range(len(data) - input_length - output_length):
        x.append(data[i:i + input_length])
        y.append(data[i + input_length:i + input_length + output_length])
    return np.array(x), np.array(y)

# 입력 및 출력 데이터 생성
x, y = create_sequences(values, sequence_length, predict_length)

# TensorFlow Dataset으로 변환
dataset = tf.data.Dataset.from_tensor_slices((x, y))
dataset = dataset.batch(batch_size).shuffle(1000)

# 모델 정의 (LSTM)
model = Sequential([
    LSTM(hidden_units, return_sequences=True, input_shape=(sequence_length, 1)),
    Dropout(0.2),  # Dropout 레이어 추가
    LSTM(hidden_units, return_sequences=False),  # 두 번째 LSTM 레이어 추가
    Dense(hidden_units, activation='relu'),  # Dense 레이어 추가
    Dense(predict_length)  # 최종 출력 (30분 예측)
])

# model = Sequential([
#     GRU(hidden_units, return_sequences=True, input_shape=(sequence_length, 1)),
#     Dropout(0.2),  # Dropout 레이어 추가
#     GRU(hidden_units, return_sequences=False),  # 두 번째 GRU 레이어 추가
#     Dense(hidden_units, activation='relu'),  # Dense 레이어 추가
#     Dense(predict_length)  # 최종 출력 (30분 예측)
# ])

# 모델 컴파일
model.compile(optimizer='adam', loss='mse')

# 데이터 학습
history = model.fit(dataset, epochs=num_epochs)

# 모델 저장
model.save("lstm_model.h5")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model

# 데이터 로드 및 준비
# _value 열만 사용, 데이터는 CSV 파일로 가정
data = pd.read_csv("simulation/total_data.csv", parse_dates=["_time"])
values = data["_value"].values  # numpy array로 변환

# 데이터 준비 (60개 입력 -> 10개 예측)
sequence_length = 60
predict_length = 30

# 데이터 슬라이싱
input_data = values[:sequence_length]  # 60개 데이터
actual_future = values[sequence_length:sequence_length + predict_length]  # 실제 값 (검증용)

# 모델 로드
model = load_model("lstm_model.h5")

# 예측 수행
input_data_tensor = np.expand_dims(input_data, axis=0)  # 배치 차원 추가 (1, 60)
input_data_tensor = np.expand_dims(input_data_tensor, axis=-1)  # (1, 60, 1)
predicted_future = model.predict(input_data_tensor).flatten()  # 예측 결과 (1, 30) -> (30,)

# 결과 시각화
plt.figure(figsize=(23, 8))

# 실제 데이터 (원본)
plt.plot(values[:90], label='Original Data', color='orange')

# 예측 데이터 표시
start_index = sequence_length
end_index = start_index + predict_length
plt.axvline(sequence_length, color='red', linestyle='--', label='Prediction Start')
plt.plot(range(start_index, end_index), predicted_future, label='LSTM Forecast', color='blue', linestyle='dotted')

# 그래프 설정
plt.grid()
plt.legend()
plt.title("LSTM Prediction vs Original Data")
plt.show()
```
