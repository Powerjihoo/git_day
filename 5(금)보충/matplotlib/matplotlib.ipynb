```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# object로 쓰면 더 효율적으로 사용할 수 있다.
fig = plt.figure(figsize=(7,7),facecolor='linen')
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)

ax1.plot([2,2,3])
ax2.plot([3,2,1])
ax3.plot([2,2,2])
```
