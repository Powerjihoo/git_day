```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# 랜덤 시드 설정
np.random.seed(42)

# 1시간 간격으로 2일치 데이터 생성
date_range = pd.date_range(start='2022-01-01', end='2022-01-03', freq='H')

# 큰 패턴 없이 랜덤하게 변동하는 데이터 생성
base_temp = 37 + np.random.normal(0, 0.1, len(date_range))

# 데이터프레임 생성
df = pd.DataFrame({'Date': date_range, 'Temperature': base_temp})
df.set_index('Date', inplace=True)

# ARIMA 모델 학습
model = ARIMA(df['Temperature'], order=(1, 2, 1))
model_fit = model.fit()

# 예측 기간 설정 (다음 1시간 예측)
forecast_steps = 50
forecast = model_fit.forecast(steps=forecast_steps)

# 예측 결과를 데이터프레임으로 변환
forecast_index = pd.date_range(start=df.index[-1] + pd.Timedelta(hours=1), periods=forecast_steps, freq='H')
forecast_df = pd.DataFrame({'Forecast': forecast}, index=forecast_index)

# 결과 시각화
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Temperature'], label='Observed Temperature', color='blue')
plt.plot(forecast_df.index, forecast_df['Forecast'], label='Forecasted Temperature', color='red', linestyle='--')
plt.title('Temperature Forecast using ARIMA')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# 랜덤 시드 설정
np.random.seed(42)

# 1시간 간격으로 2일치 데이터 생성
date_range = pd.date_range(start='2022-01-01', end='2022-01-03', freq='H')

# 큰 패턴 없이 랜덤하게 변동하는 데이터 생성
base_temp = 37 + np.random.normal(0, 0.1, len(date_range))

# 데이터프레임 생성
df = pd.DataFrame({'Date': date_range, 'Temperature': base_temp})
df.set_index('Date', inplace=True)

# ARIMA 모델 학습
model = ARIMA(df['Temperature'], order=(1, 2, 1))
model_fit = model.fit()

# 예측 기간 설정 (다음 50시간 예측)
forecast_steps = 50
forecast = model_fit.forecast(steps=forecast_steps)

# 예측 결과를 데이터프레임으로 변환
forecast_index = pd.date_range(start=df.index[-1] + pd.Timedelta(hours=1), periods=forecast_steps, freq='H')
forecast_df = pd.DataFrame({'Forecast': forecast}, index=forecast_index)

# Threshold 설정
threshold = 0.05
forecast_df['Upper Bound'] = forecast_df['Forecast'] + threshold
forecast_df['Lower Bound'] = forecast_df['Forecast'] - threshold

# 결과 시각화
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Temperature'], label='Observed Temperature', color='blue')
plt.plot(forecast_df.index, forecast_df['Forecast'], label='Forecasted Temperature', color='red', linestyle='--')
plt.fill_between(forecast_df.index, forecast_df['Lower Bound'], forecast_df['Upper Bound'], color='gray', alpha=0.2, label='Forecast ±0.05°C')
plt.title('Temperature Forecast using ARIMA with Threshold')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid()
plt.show()
```
