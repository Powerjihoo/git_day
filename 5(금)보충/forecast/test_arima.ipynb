```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
```

```python
value = [1,2,3,4,5,4,3,2,1]*100
```

```python
model = ARIMA(value[:-10], order=(2, 1, 2)).fit()
forecast = model.forecast(steps=10)
```

```python
plt.figure(figsize=(20,5))
plt.plot(value[-10:], label="real")
plt.plot(forecast, label="predict")
plt.legend()
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
import matplotlib.pyplot as plt

# ARIMA 모델 학습
model = ARIMA(df, order=(2, 1, 2))
model_fit = model.fit()

# 예측
forecast_steps = 60
forecast = model_fit.forecast(steps=forecast_steps)

# 실제 데이터와 예측 데이터 시각화
plt.figure(figsize=(10, 5))
plt.plot(df, label='Actual Data')
plt.plot(np.arange(df['time'].iloc[-1] + 1, df['time'].iloc[-1] + 1 + forecast_steps), forecast, label='Forecast', color='red')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
